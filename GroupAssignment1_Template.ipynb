# Uber Business Model 

### what is your idea?

-- Our idea is about Uber. Uber is a inexpensive and safe taxi service. The uber company hires drivers to pick up customers. 

### what makes it unique?

-- When working with Uber, there are flexible schedules. Employees can decide if they want to work part-time or full-time. Their workers are like a contractor; they are just entering in an agreement to pick up customers. 

### is this a brand-new product or business, or are you borrowing from another business?

-- We are borrowing from another business, which is Uber. We are not adding anything to the buisness model at this time. 

### describe a typical business interaction or a day at your business.

--  Customers go on the Uber App and ask for a ride. The Uber drivers decide if they want to pick them up. After the Uber drivers picked the customers, they will drive them to the location. Last, customers will pay for and rate their Uber drivers through the app.



### list your 5 entities/tables and for each, provide details about what data will be stored in the entities/tables.

1. Customers. This entity will store the customers first name, last name, and give them a unique app ID number.

-- Customers schedule a ride. 

2. Drivers. This entity will store the drivers first name, last name, and give the driver a unique driver number.

-- Drivers pick up the customers. 

3. App. This entity will store the ride scheduler, the payment tool, and the rating system.

-- Customers and Drivers use the Uber App.

4. Locations. This entity will store the address, name, and distance from the customer.

-- Customers and Drivers go to a destination. 

5. Company. This entity will store the customer relations, marketing, and law team attributes. 

-- Company sends the drivers to pick up the customers.
