Above is my Pet Shop ER Diagram. Let's begin with the relationship between pets and foods: one bag of dog/cat food can feed three dogs/cats, and one dog/cat eats one kind of dog/cat food. The relationship between food transactions and food: one transaction can print two foods, while zero food can be printed in one hundred transactions. Next, the relationship between food transactions and customers: one transaction is purchased by one customer, while one customer can buy five transactions. For the relationship between pets and enclosures: one enclosure can keep two pets, while one pet can be kept in one enclosure. Last but not least, the relationship between pets and pet transactions: zero pets can be printed on one hundred transactions, while one transaction can print one pet.
